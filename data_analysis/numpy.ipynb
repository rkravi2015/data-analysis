```python
#[apple,google,microsoft]
bob = np.array([1000, 1000, 1000])
alice = np.array([500, 2000, 500])
porfolios = np.array([bob, alice])
```

```python
porfolios
```

```python
changes = np.array([[1.2, 0, 0], #apple's stock incre by 20%
                   [0, 0.95, 0], #google decreased by 5%
                   [0, 0, 1],#same for microsoft
                   [0.2, -0.05, 0]#total stock change
                   ] )
```

```python
np.matmul(changes, porfolios)
```

```python
porfolios.transpose().shape
```

```python
np.matmul(changes, porfolios.transpose()).transpose()
```
